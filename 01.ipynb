## Prompt:
make the following code in 01.ipynb: load the 01.test.txt file and parse it line by line into two arrays of numbers, one for each column in the data. Then sort each array ascending. Then get the difference between each corresponding row, and then sum the differences



```python
# Read and process the data
with open('01.txt', 'r') as file:
    lines = file.readlines()

# Parse into two arrays
array1 = []
array2 = []

for line in lines:
    if line.strip():  # Skip empty lines
        nums = line.strip().split()
        array1.append(int(nums[0]))
        array2.append(int(nums[1]))

# Sort both arrays
sorted_array1 = sorted(array1)
sorted_array2 = sorted(array2)

# Calculate differences
differences = [abs(a - b) for a, b in zip(sorted_array1, sorted_array2)]

# Sum the differences
total_difference = sum(differences)

# print(f"Original arrays: \nArray 1: {array1}\nArray 2: {array2}")
# print(f"\nSorted arrays: \nArray 1: {sorted_array1}\nArray 2: {sorted_array2}")
# print(f"\nDifferences: {differences}")
print(f"Sum of differences: {total_difference}")
```
