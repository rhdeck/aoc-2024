```python
import numpy as np

def load_data(filename):
    # Read the file and convert to numpy array
    with open(filename, 'r') as f:
        lines = f.readlines()
    
    # Remove any trailing whitespace and empty lines
    lines = [line.strip() for line in lines if line.strip()]
    
    # Convert to 2D numpy array
    # Each line is converted into a list of characters
    try:
        data = np.array([list(line) for line in lines])
        return data
    except Exception as e:
        print(f"Error converting data to numpy array: {e}")
        return None

filename = "04.txt"
data = load_data(filename)
if data is not None:
    print("Data shape:", data.shape)
    # print("Data:")
    # print(data)
```

```python
def print_matrix(matrix: List[List[str]]) -> None:
    """
    Helper function to print a 2D matrix.
    """
    for row in matrix:
        print(''.join(row))

print_matrix(data)
```

```python
from typing import List
import math
import re

rotData = data

def rotate_45_degrees(matrix: List[List[str]]) -> List[List[str]]:
    """
    Rotates a 2D matrix of strings by 45 degrees clockwise.
    
    Args:
        matrix: A 2D list of strings to be rotated
        
    Returns:
        A new 2D list representing the rotated matrix
    """
    # if not matrix or not matrix[0]:
    #     return []
    
    rows = len(matrix)
    cols = len(matrix[0])
    
    # Calculate dimensions of the rotated matrix
    new_width = rows + cols - 1
    new_height = rows + cols - 1
    
    # Create empty rotated matrix filled with spaces
    rotated = [[' ' for _ in range(new_width)] for _ in range(new_height)]
    
    # Calculate center point for rotation
    center_x = new_width // 2
    center_y = new_height // 2
    
    # For each point in the original matrix
    for y in range(rows):
        for x in range(cols):
            # Calculate the offset from the center of the original matrix
            dx = x - (cols - 1) / 2
            dy = y - (rows - 1) / 2
            
            # Apply 45-degree rotation transformation
            # x' = x*cos(θ) - y*sin(θ)
            # y' = x*sin(θ) + y*cos(θ)
            # where θ = 45° = π/4
            angle = math.pi / 4  # 45 degrees in radians
            new_x = dx * math.cos(angle) - dy * math.sin(angle)
            new_y = dx * math.sin(angle) + dy * math.cos(angle)
            
            # Scale the coordinates (45° rotation causes scaling)
            scale = math.sqrt(2)
            new_x *= scale
            new_y *= scale
            
            # Convert back to matrix coordinates and round to nearest integer
            matrix_x = round(new_x + center_x)
            matrix_y = round(new_y + center_y)
            
            # If the new coordinates are within bounds, copy the value
            if (0 <= matrix_x < new_width and 0 <= matrix_y < new_height):
                rotated[matrix_y][matrix_x] = matrix[y][x]
    
    return rotated




pattern = r'XMAS'
xmases = 0
for i in range(8):
    print("starting with iteration", i)
    # print_matrix(rotData)
    for row in rotData:
        string = ''.join([r for r in row if (not r.isspace())])
        # print("printing a string")
        # print(string)
        if len(string.strip()) == 0:
            continue
        matches = re.findall(pattern, string.strip())  
        xmases += len(matches)
    print("rotating", len(rotData), sum(len(row) for row in rotData)), 
    rotData = rotate_45_degrees(rotData)
    # rotData = [re.sub(r'[^a-zA-Z]', '', ''.join(s)).split() for s in rotData if (''.join(s)).strip() != '']
    # print_matrix(rotData)

print(xmases)

# rotated_matrix = rotate_45_degrees(data)
# print_matrix(rotated_matrix)
```
